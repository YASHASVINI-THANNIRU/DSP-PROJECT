# Choosing Independent and dependent values

```python
#predicting Bitcoin price for next 14 days
projection=14
df['prediction_BTC']=df[['Adj Close (BTC)']].shift(-projection)
df
```

```python
X=np.array(df[['Adj Close (BTC)','Volume (BTC)']])
X=X[:-projection]
X
```

```python
Y=df['prediction_BTC'].values
Y=Y[:-projection]
Y
```

# Training and Testing

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.15)
```

# Evalution

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
linreg=LinearRegression()
linreg.fit(x_train,y_train)
linreg_score_train=linreg.score(x_train,y_train)
linreg_score_test=linreg.score(x_test,y_test)
print("Testing Accuracy",linreg_score_test)
print("training accuracy",linreg_score_train)
```

```python
neigh = KNeighborsRegressor(n_neighbors=2)
d=neigh.fit(x_train, y_train)
KNN_score_train=d.score(x_train,y_train)
KNN_score_test=d.score(x_test,y_test)
print("Testing Accuracy",KNN_score_test)
print("training accuracy",KNN_score_train)
```

```python
rf = RandomForestRegressor()
a=rf.fit(x_train,y_train)
RF_score_test=a.score(x_test,y_test)
RF_score_train=a.score(x_train,y_train)
print("Testing Accuracy",RF_score_test)
print("training accuracy",RF_score_train)
```

```python
dt = tree.DecisionTreeRegressor()
b=dt.fit(x_train, y_train)
DT_score_test=b.score(x_test,y_test)
DT_score_train=b.score(x_train,y_train)
print("Testing Accuracy",DT_score_test)
print("training accuracy",DT_score_train)
```

```python
data=[linreg_score_test,KNN_score_test,RF_score_test,DT_score_test]
lab=['LinearRegressor','KNeighborsRegressor','RandomForestRegressor','DecisionTreeRegressor']
plt.barh(lab,data)
plt.show()
```
